```python
df = pd.read_csv("C:/Users/dbda.STUDENTSDC/Downloads/output.csv")
df
```

```python
# from datetime import datetime


df['Start date'] = pd.to_datetime(df['Start date'], errors='coerce')
df['Experience'] = (pd.Timestamp.today() - df['Start date']).dt.days / 365.25
df['Experience'] = df['Experience'].round(0)  # Optional: round to 1 decimal
df
```

```python
bin_edges = [12, 13, 14, 15, 16, 17]  # Define explicit bin edges
plt.hist(df.Experience, bins=bin_edges, align='mid')  # 'mid' aligns bars with bin edges
plt.show()
```
